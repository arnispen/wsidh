```python
import matplotlib.pyplot as plt
```

```python
lines = open("wsidh_dump.csv").read().splitlines()

blocks = {}
current = None
```

```python
for line in lines:
    line = line.strip()
    if not line:
        continue  # skip empty lines

    # Header line
    if line.startswith("# "):
        current = line[2:]
        blocks[current] = []
        continue

    # Skip any line that isn't comma-separated numbers
    if current is None:
        continue

    try:
        nums = [int(x) for x in line.split(",") if x.strip() != ""]
        blocks[current] = nums
    except ValueError:
        # Line isn't valid data (like "WSIDH KEM test: SUCCESS")
        continue
```

```python
# Plot everything
for name, coeffs in blocks.items():
    plt.figure(figsize=(14, 3))
    plt.plot(coeffs)
    plt.title(name)
    plt.grid(True)
    plt.show()
```
